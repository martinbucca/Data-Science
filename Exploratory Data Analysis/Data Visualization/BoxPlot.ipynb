```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
```

```python
def to_kb(size):
    if size.endswith('M'):
        return float(size[:-1]) * 1000
    elif size.endswith('k'):
        return float(size[:-1])

```

```python
df = pd.read_csv('/content/drive/MyDrive/Organizacion de datos/googleplaystore.csv')
dataframe_limpio = df[df['Size'].notna() & (df['Size'] != 'Varies with device') & df['Size'].str.match(r'^\d+(\.\d+)?[Mk]$')]
dataframe_limpio['Size'] = dataframe_limpio['Size'].apply(to_kb)

plt.figure(figsize=(12, 4))
sns.boxplot(data=dataframe_limpio, x='Category', y='Size', width=0.8)
plt.title('Distribucion del tamaño de las apps por categoria en kB')
plt.xlabel('Categoría de la App')
plt.ylabel('Tamaño (en kB)')
plt.xticks(rotation=45, ha='right')
plt.subplots_adjust(bottom=0.2, right=0.9)

plt.show()






```
