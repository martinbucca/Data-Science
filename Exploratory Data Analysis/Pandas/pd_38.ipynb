Para cada categoría, calcular cuál es el caracter predominante en la misma. El caracter predominante en una categoria esta dado por el caracter que es dominante en mayor número de apps, por ej para aaaaaaax, xxb, xxs el caracter dominante de ap1 es a, pero para las otras 2 x, por lo que x es el caracter dominante en la categoría. Si no hay caracter dominante, es el primer caracter. (⭐⭐)

Defino que distinga entre mayusculas y minusculas, ya que los considero "caracteres distintos". Decido ignorar el espacio ' ' como un caracter que cuenta como predominante.

```python
def caracter_dominante(app):
  caracteres_de_app = {}
  for caracter in app:
    if caracter == ' ':
      continue
    if caracter not in caracteres_de_app:
      caracteres_de_app[caracter] = 1
    else:
      caracteres_de_app[caracter] += 1
  caracter_predominante = max(caracteres_de_app, key=caracteres_de_app.get)
  return caracter_predominante

```

```python
import pandas as pd
path = '/content/drive/MyDrive/Organizacion de datos/TP1 - Pandas, Spark y Visualización de datos/googleplaystore.csv'
apps_df = pd.read_csv(path, usecols=['App', 'Category'])
apps_df.dropna(subset=['App'], inplace=True)
apps_df['Dominant Character'] = apps_df['App'].apply(caracter_dominante)
```
