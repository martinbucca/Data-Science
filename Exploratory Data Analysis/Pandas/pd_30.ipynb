¿Cuál es la aplicación gratis con mayor ratio de reviews positivas? (⭐⭐)


```python
import pandas as pd
path_google_playstore = '/content/drive/MyDrive/Organizacion de datos/TP1 - Pandas, Spark y Visualización de datos/googleplaystore.csv'
path_reviews = '/content/drive/MyDrive/Organizacion de datos/TP1 - Pandas, Spark y Visualización de datos/googleplaystore_user_reviews.csv'

apps_df = pd.read_csv(path_google_playstore, usecols= ['App', 'Type'])
reviews_df = pd.read_csv(path_reviews, usecols= ['App', 'Sentiment'])
free_apps = apps_df[apps_df['Type'] == 'Free']
free_apps_with_reviews = pd.merge(free_apps, reviews_df, on='App')
positive_reviews_ratio = free_apps_with_reviews.groupby('App')['Sentiment'].apply(lambda x: (x == 'Positive').mean()).reset_index(name='Positive Reviews Ratio')
app_with_best_ratio_idx = positive_reviews_ratio['Positive Reviews Ratio'].idxmax()
positive_reviews_ratio.loc[app_with_best_ratio_idx, 'App']
```
